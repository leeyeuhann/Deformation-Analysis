```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv('Test data.csv')
Data
```

```python
Data['t']=Data['t'].ffill()
Data['Cell']=Data['Cell'].ffill()
Data['Cell']=Data['Cell'].astype('int64')
Data = Data.set_index(['t','Cell','Point'])
Data.index.names
```

```python
Data
```

```python
timescale = len(Data.groupby(level=0))
cellnum = len(Data.groupby(level=1))
posnum = len(Data.groupby(level=2))
```

```python
def flatten(t):
    return [item for sublist in t for item in sublist]
```

```python
# number of columns
num_columns = (len(Data.columns)) - 3
num_columns
max_intense_list = []
z_intense_list = []
datalist = [] 
# for loop to iterate over time
# for loop to iterate over cells
# for loop to iterate data
for t,new_df in Data.groupby(level=[0,1]):
    differences = new_df.diff(periods = -2, axis = 1)
    max_intensity = differences.max().max()
    max_intense_list.append(max_intensity)
    filter = (differences == max_intensity).any()
    z_intense = differences.loc[: , filter]
    z_intense = z_intense.loc[differences.isin([max_intensity]).any(axis=1)]
    z_intense.reset_index(inplace=True)
    z_intense = flatten(z_intense.to_numpy('float').tolist())
    print(max_intensity)
    print(z_intense)
    datalist.append(z_intense)

```

```python
Processed_Data = pd.DataFrame(datalist,columns=['Time','Cell','Point','Max Intensity'])
Processed_Data
```
